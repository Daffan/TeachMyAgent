```python
with open("/home/users/zifanxu/TeachMyAgent/TeachMyAgent/students/data/profiling_benchmark_stumps_ALP-GMM_criteria_2/profiling_benchmark_stumps_ALP-GMM_criteria_2_s16/eval_result_00825.json", "r") as f:
    eval_rewards = json.load(f)

stump_heights = [d["stump_height"] for d in eval_rewards]
obstacle_spacings = [d["obstacle_spacing"] for d in eval_rewards]
rewards = [d["reward"] for d in eval_rewards]
mastered = [d["reward"] > 230 for d in eval_rewards]
print("mastered task %d" %(sum(mastered)))
plt.scatter(stump_heights, obstacle_spacings, c=rewards)
plt.colorbar()
plt.show()

plt.scatter(stump_heights, obstacle_spacings, c=mastered)
plt.colorbar()
plt.show()
```
